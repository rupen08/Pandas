```python
## how = "outer"

week1.merge(right = week2, how = "outer", on = "Customer ID", suffixes = ["- week 1", "- week 2"])
```

```python
## if outer matches same value based on customer id, It will give only one row and also returns 
## the values which only present for once
## so, it gives shared and unshared values but shared values saved in single raw So, there will be less rows than 
## simple concat or append
## NaN means did not come in that week
```

```python
## it we want the indicator, whether it is merge or left or right
```

```python
a = week1.merge(right = week2, how = "outer", on = "Customer ID", suffixes = ["- week 1", "- week 2"], indicator = True)
a
```

```python
import pandas as pd

week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv", index_col = "ID")
food = pd.read_csv("Restaurant - Foods.csv", index_col = "Food ID")
satis = pd.read_csv("Restaurant - Week 1 Satisfaction.csv")
```

```python
## targeting week1 as left

week1.head(3)
```

```python
customers.head(3)
```

```python
food.head(3)
```

```python
## left = week1
## right = customer

sales_week1 = week1.merge(customers, how = "left", left_on = "Customer ID", right_index = True, sort = True)

## advantage is the multi merge with no repeatation in ID column like before
```

```python
sales_week1.head(3)
```

```python
sales_week1_final = sales_week1.merge(food, how = "left", left_on = "Food ID", right_index = True, sort = True)
```

```python
sales_week1_final.head(3)
```
