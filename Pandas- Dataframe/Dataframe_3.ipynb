## sample method

```python
import pandas as pd
```

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
```

```python
bond.sort_index(inplace = True)
```

```python
bond.head(3)
```

```python
## sample method gives random data at random order even if they are sorted
## n is how many data you want
## gives 2 raws and all columns

bond.sample(n = 2)
```

```python
## farc means what percentage of data you want from the file, divide that number by 100

bond.sample(frac = 0.25)  ## 25% data is asked
```

```python
## gives random 4 axis and all raws

b = bond.sample (n = 4, axis = 1)
```

## nlargest and nsmallest method

```python
## used when we need large and small amount from a collumn for n numbers of raws
```

```python
import pandas as pd

bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_values("Film", inplace = True)


bond.head(3)
```

```python
bond.nlargest(n = 4, columns = "Box Office" )

## this will also give in large to small number sorting
```

```python
bond.nsmallest(n = 4, columns = "Bond Actor Salary")
```

```python
## or the other way
## Series output

(bond["Box Office"].nlargest(4))
```
