## fillna method

```python
nba.head(3)
```

```python
## fillna method fills NaN values with given value in paranthesis 
nba.fillna(0).head(3)
```

```python
## the problem is, it will replcae all the NaN with the 0.
## it is okay if the column is float64 or int64 but object.
## so, the better way is to go with series 

nba["Salary"].head(3)
```

```python
nba["Salary"].fillna(0).head(3)
## proof
```

```python
nba["College"].head(5)
```

```python
nba["College"].fillna("Never Attended").head(5)
## but this will not change the original DataFrame column
```

```python
nba["College"].fillna("Never Attended", inplace = True)
```

## category data type

```python
## if the data has millions of raws but for specific raw the values are unique or not repeteative
## 20k raws but the gender column has only two arguments, called category data

## nunique() gives max unique values for the column, no. of rwas not metter

nba["Position"].nunique()
```

```python
## change these type of columns into category data type, this will save significant memory
```

```python
nba["Position"] = nba["Position"].astype("category")
```

```python
nba["Team"] = nba["Team"].astype("category")
```

```python
nba.info()
```

```python
## from 35 to 32 and 32 to 27 kb
```
