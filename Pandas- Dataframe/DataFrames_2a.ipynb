## .duplicated()

```python
df.sort_values("First Name", inplace = True)
```

```python
df.head()
```

```python
## now aaron is repeated 4times. 

mask5 = df["First Name"].duplicated( keep = "first")

df["First Name"].duplicated( keep = "first")

## this will keep first value as a unique and then onwards duplicate.
## that is why there is False et first because its not duplicate but rest is.
```

```python
df[mask5] ##101 is unique ## gives DUPLICATE  list
```

```python
mask6 = df["First Name"].duplicated( keep = "last")
 ## this will starts count from bottom so last same name will have unique as that is the first one from the last

df[mask6] ##937 is unique
```

```python
## now let say, if the name is repeated ever, all of them i want to marrk duplicated 
    
df["First Name"].duplicated( keep = False)
    ## all true means all duplicate
```

```python
mask7 = df["First Name"].duplicated( keep = False)
df[mask7]
## all duplicate
```

```python
## now let say you want to print only values which are unique (only one time)

mask8 = ~ df["First Name"].duplicated( keep = False)  ## there is no True in keep
df[mask8]

## data which are unique
```
