## .apply() method for dataframe for columns

```python
import pandas as pd
```

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
```

```python
bond.head(3)
```

```python
def int_to_str(number):
    return str(number) + " " + "MILLIONS"
```

```python
bond["Box Office"] = bond["Box Office"].apply(int_to_str)
```

```python
bond
```

```python
bond["Bond Actor Salary"].fillna(0, inplace = True)
```

```python
bond.head(3)
```

```python
bond["Bond Actor Salary"] = bond["Bond Actor Salary"].apply(int_to_str)
```

```python
bond.head(3)
```

```python
## now if the columns are more than one what can be done is, one can provide columns name in separate variable and run for
## loop.

bond1 = pd.read_csv("jamesbond.csv", index_col = "Film")
bond1.sort_index(inplace = True)
```

```python
bond1.head(3)
```

```python
## before that just fill NaN values in salary col.

bond1["Bond Actor Salary"].fillna(0, inplace = True)
```

```python
def int_to_str(number):
    return str(number) + " " + "MILLIONS"
```

```python
cols = ["Box Office", "Budget", "Bond Actor Salary"]
```

```python
for i in cols:
    bond1[i] = bond1[i].apply(int_to_str)
```

```python
bond1.head(3)

## hence prooved!
```

 ## .apply() method for raw values in dataframe

```python
bond2 = pd.read_csv("jamesbond.csv", index_col = "Film")
bond2.sort_index(inplace = True)
```

```python
bond2.head()
```

```python
def movie_review(row):
    
    ## assigning which column want to work with for each row
    
    actor = row[1]
    box_office = row[3]
    
    if actor == "Daniel Craig":
        return "Best Movie"
    elif box_office > str(500):
        return "Watchable"
    elif actor == "Sean Connery" and box_office > str(100):
        return "Good Watch"
    else:
        return "No Idea!"
    
## the code is for the each row in the data uses only two columns as a target
```

```python
bond.apply(movie_review, axis = "columns")

## axis = columns is important because here we are going row to row with more than one column
```
