## rank method

```python
## use to give rank the data column
## let say i want to know who has highest salary
nba = pd.read_csv("nba.csv").dropna(how = "all")
nba["Salary"].fillna(0, inplace= True)
nba.head()
```

```python
nba["Salary"].rank().head()
```

```python
## we want highest paid player at number 1 but panda rank is giving the reverse having ascending= True by defualt 
## and that also in float
## two problems

nba["Salary"].rank(ascending = False).astype("int")
```

```python
nba["Salary rank"] = nba["Salary"].rank(ascending = False).astype("int")
nba.sort_values("Salary", ascending = False, inplace= True)
```

```python
nba.head()
```

```python
nba["College"] = nba["College"].fillna("Not Attended")
```
