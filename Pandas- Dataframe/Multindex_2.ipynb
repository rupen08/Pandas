```python
import pandas as pd
```

```python
world = pd.read_csv("worldstats.csv", index_col = ["country", "year"])
```

```python
world
```

```python
## stack method gives whole data in series (one dimenstional)
## 3 index as a 1 and 1 column

s = world.stack()
s
```

```python
## to convert into data_frame
world.stack().to_frame()
```

```python
s
```

### unstack() method

```python
## starts with most inner layer (right side) and made it in column

s.unstack()
```

```python
## we can unstack till we have last inner layer left
```

```python
s.unstack().unstack()
```

```python
s.unstack().unstack().unstack()
```

## .pivot() method

```python
import pandas as pd
```

```python
sales = pd.read_csv("salesmen.csv", parse_dates = ["Date"])
```

```python
sales.head(3)
```

```python
sales.pivot(index = "Date", columns = "Salesman", values = "Revenue").head()
```

## .pivot_table()

```python
## something interesting
```

```python
sales.pivot_table(values = "Revenue", index = ["Date","Salesman"], aggfunc = "sum")
```

```python
sales.pivot_table(values = "Revenue", index = ["Date"], aggfunc = "sum")
```

```python
sales.pivot_table(values = "Revenue", index = ["Salesman"], aggfunc = "sum")
```
